```python
# NYT reference: https://www.nytimes.com/interactive/2021/07/27/business/economy/covid-parenting-work-time.html

import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

image.png

```python
data = {'Alone': [6, 7],
        'With people outside household': [4.3, 2.7],
        'With household members only': [3.9, 4.4]}
df = pd.DataFrame(data = data,
                  index = [2019, 2020])
df
```

```python
df.plot()
plt.title('A lonely year')
plt.xlabel('Average time spent per day during waking hours\nMay through December in 2020 vs 2019', labelpad = 20)
plt.show()
```

```python
dark_gray = (.2, .2, .2)

fig = plt.figure(figsize = (5,7))
ax = plt.axes(facecolor = (.95, .96, .97))

df.plot(ax = ax, marker = 'o', markersize = 8, linewidth = 3, clip_on = False)

ax.set_xlim(2019,2020)
ax.ticklabel_format(axis = 'x', useOffset = False)
ax.xaxis.set_ticks([2019,2020])
ax.xaxis.set_tick_params(length = 0, labeltop = True, labelbottom = False, labelcolor = dark_gray)

ax.get_xticklabels()[0].set_ha('left')
ax.get_xticklabels()[1].set_ha('right')

ax.legend().set_visible(False)

for key in 'left', 'right', 'top':
  ax.spines[key].set_visible(False)

# y ticks and grid
ax.yaxis.grid(True, color = 'white', linewidth = 2)

yticks = [0,2,4,6,8]
ax.set_yticks(yticks)
ax.set_yticklabels([])
ax.yaxis.set_tick_params(length = 0)

ax.text(2019.5, 0, 'No time', va = 'bottom', ha = 'center', color = dark_gray)
for ytick in yticks[1:]:
  ax.text(2019.5, ytick + .01, '{:} hours'.format(ytick), va = 'bottom', ha = 'center', color = dark_gray)


# labels
left_text_style = dict(ha = 'right', fontname = 'Liberation Sans', fontweight = 'bold', size = 12)
right_text_style = dict(ha = 'left', size = 11, fontname = 'Liberation Sans', color = dark_gray, va = 'center')

# Alone
x = 2019 - .02
ax.text(x, df.loc[2019, 'Alone'], 'Alone', color = 'C0', va = 'center', **left_text_style)
ax.text(2020 + .03, df.loc[2020, 'Alone'], '+57 min.', **right_text_style)

# people outside household
ax.text(x, df.loc[2019, 'With people outside household'], 'With people\noutside household', color = 'C1', va = 'bottom', **left_text_style)
ax.text(2020 + .03, df.loc[2020, 'With people outside household'], '-1 hour and 33min.', **right_text_style)


# With household members only
ax.text(x, df.loc[2019, 'With household members only'], 'With\nhousehold\nmembers only', color = 'C2', va = 'top', **left_text_style)
ax.text(2020 + .03, df.loc[2020, 'With household members only'], '+31 min.', **right_text_style)


ax.set_title('A lonely year', size = 15, pad = 60, fontname = 'Liberation Sans', fontweight = 'bold')

subtitle = 'Average time spent per day during waking hours, May through\nDecember in 2020 vs. 2019'
ax.text(0.5, 1.08, s = subtitle, transform = ax.transAxes, ha = 'center', fontname = 'Liberation Sans', color = dark_gray)

plt.show()
```
